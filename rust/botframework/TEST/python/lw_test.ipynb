```python
import pandas as pd
import numpy as np
import rbot.lightweight as lw
```

```python
length = 10000

df = pd.DataFrame(
    np.random.choice([-1, 0, 1], size=length).cumsum(0).T,
    index=pd.date_range('2010-01-01', periods=length, freq='10S'),
    columns=['price']
)

df['ma'] = df['price'].rolling(100).mean()
df = df.dropna()
```

```python
lw.plot(
    data=df[['price', 'ma']],
    config=[
        {'name': 'price', 'fn': 'addLineSeries', 'style': {}},
        {'name': 'ma', 'fn': 'addLineSeries', 'style': {'color': 'orange', 'lineWidth': 2}}
    ]
)
```

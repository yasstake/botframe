```python
import numpy as np

# period先を予測する　＝　diffはマイナス方向。
def log_return(series, periods=1):
    return np.log(series).diff(periods=-periods)
```

```python
df = rbot.array_to_df(bb_btcusd.ohlcv(0, 0, 60))
```

```python
df
```

```python
df['lr01'] = log_return(df['close'], 1)
df['lr02'] = log_return(df['close'], 2)
df['lr03'] = log_return(df['close'], 3)
df['lr04'] = log_return(df['close'], 4)
df['lr05'] = log_return(df['close'], 5)
df['lr10'] = log_return(df['close'], 10)
df['lr15'] = log_return(df['close'], 15)
df
```
